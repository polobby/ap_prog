```python
from sklearn import svm,linear_model, naive_bayes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python


#les variables "month" et "day" (aussi "X" et "Y" ?), on fait du one hot encoding pour les convertir en variable numériques
#car le svm ne fonctionne que sur des données numériques

# Séparez les colonnes numériques et catégorielles
numerical_cols = data.select_dtypes(include=['int', 'float']).columns
categorical_cols = data.select_dtypes(include=['object']).columns

preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', numerical_cols),  # conserve les colonnes numériques
        ('cat', OneHotEncoder(), categorical_cols)  # applique OneHotEncoder aux colonnes catégorielles
    ])

# Appliquez la transformation
data_encoded = pd.DataFrame(preprocessor.fit_transform(data), columns=list(numerical_cols) + list(preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols)))


#on sauvegarde les noms des colonnes dans un dictionnaire, car on va passer en numpy et les array numpy ne conservent pas les noms de colonnes
column_names = data_encoded.columns.tolist()
dico_columns = {variable: indice for indice, variable in enumerate(column_names)}

#passage à un numpy array
data_array = np.array(data_encoded)


X = data_array[:,:-1]
y = data_array[:,-1]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)


mod_SVM1 = svm.SVC(kernel="linear")
mod_SVM2 = svm.SVC(gamma = 2)

mod_SVM1.fit(X_train, y_train)
mod_SVM2.fit(X_train, y_train)

yhat1 = mod_SVM1.predict(X_test)
yhat2 = mod_SVM2.predict(X_test)


```
